Happy pi-day!

```python
from matplotlib import pyplot as plt
import random
import numpy as np

%matplotlib inline
```

```python
#%%timeit
est = []
in_circle = 0
xs = []
ys = []
ins = []

for i in range(1000000):
  x = random.random()
  y = random.random()
  radius = (x**2 + y**2)**0.5

  if radius <= 1:
    in_circle += 1
    ins.append(1)
  else:
    ins.append(0)
  
  xs.append(x)
  ys.append(y)
  est.append(4*in_circle/len(xs))

fig,axs = plt.subplots(ncols=2, figsize=(8,4))
axs[0].scatter(xs, ys, alpha = 0.1, c = ins)
axs[1].plot(est)
axs[1].hlines(np.pi, 0, len(est))
axs[1].set_ylim(3.12, 3.18)
```
